# Step 1: Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Step 2: Load dataset

```python
dataset=pd.read_csv("H:/corsera/globalterrorismdb_0718dist.csv")
dataset.head()
```

# Step 3: Data Cleaning

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset.head()
```

```python
dataset.rename(columns={"iyear" : "Year", "imonth" : "Month", "iday" : "Day", "country_txt" : "Country",
                   "region_txt" : "Region", "provstate" : "State", "latitude" : "Latitude", 
                   "longitude" : "Longitude", "summary" : "Summary", "attacktype1_txt" : "AttackType", 
                   "targtype1_txt" : "TargetType", "gname" : "Group", "weaptype1_txt" : "WeaponType", 
                   "nkill" : "Kills", "nwound" : "Wounded",}, inplace=True)
```

```python
dataset = dataset[["Year", "Month", "Day", "AttackType", "TargetType", "Country", "Region",
         "Latitude", "Longitude", "Summary", "State", "motive", "WeaponType", 
         "Kills", "Wounded", "Group"]]
dataset
```

# Step 4: Relationship Analysis

```python
plt.figure(figsize = (20,10))
sns.heatmap(dataset.corr(), annot = True)
```

```python
plt.figure(figsize=(9,9))
plt.title('Attack Type')
plt.pie(dataset['AttackType'].value_counts(),labels=dataset['AttackType'].value_counts().index,autopct='%1.1f%%')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,10))
dataset.groupby('Year')['Kills','Wounded'].sum().plot()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x="TargetType", data=dataset, order = dataset["TargetType"].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

x=dataset["Country"].value_counts()[:10].index

```python
plt.figure(figsize=(20,10))
sns.countplot(x="Region", data=dataset)
plt.xticks(rotation=90)
plt.title("Region v/s attack")
plt.show()
```

```python
x=dataset["State"].value_counts()[:10].index
y=dataset["State"].value_counts()[:10].values
sns.barplot(x,y)
plt.title("Top 10 States most affected by Terrorism")
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x="WeaponType", data=dataset, order = dataset["WeaponType"].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x="Month", data=dataset, order = dataset["Month"].value_counts().index)
plt.xticks(rotation=90)
plt.title("attacks on month")
plt.show()
```
