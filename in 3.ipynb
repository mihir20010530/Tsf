# Step 4: Relationship Analysis

```python
sns.heatmap(dataset.corr())
```

```python
sns.pairplot(dataset)
```

```python
sns.barplot(data=dataset, x="Ship Mode", y="Profit", hue="Region")
```

```python
sns.barplot(data=dataset, x="Segment", y="Profit", hue="Category")
```

```python
sns.barplot(data=dataset, x="Region", y="Profit", hue="Segment")
```

```python
plt.figure(figsize=(30,10))
dataset.groupby('State')['Profit'].sum().plot.bar(width=1)
plt.title("state wise profit")
```

```python
plt.figure(figsize=(30,10))
dataset.groupby('Sales')['Profit'].sum().plot()
plt.title("Sale wise profit")
```

```python
plt.figure(figsize=(30,10))
dataset.groupby('Quantity')['Profit'].sum().plot()
plt.title("Quantity of selling v/s profit")
```

```python
plt.figure(figsize=(30,20))
dataset.groupby('Discount')['Profit'].sum().plot.bar(width=1)
plt.title("Discount v/s profit")
```

```python
plt.figure(figsize=(20,10))
dataset.groupby('Region')['Profit'].sum().plot.bar()
plt.title("Region v/s profit")
```

# Step 5: Conclusion 

1) California and New york state have maximum profit.
